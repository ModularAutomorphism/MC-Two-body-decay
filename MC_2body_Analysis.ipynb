```python
# This Python code is for the analysis of data generated by MC generator for charged pion decay.
# Author: Avinandan Mondal, Dept. of Physics.

import csv
import matplotlib.pyplot as plt
import numpy as np
#from google.colab import files
```

```python
# Muon channel data

with open('muon_channel_vals_pion_momentum.csv', 'r') as file:
    reader = csv.reader(file)
    pion_mom_muon_channel_decay = []
    for row in reader:
        pion_mom_muon_channel_decay.append(float(row[0]))

with open('muon_momentum_vals.csv', 'r') as file:
    reader = csv.reader(file)
    muon_mom = []
    for row in reader:
        muon_mom.append(float(row[0]))

with open('muon_neutrino_momentum_vals.csv', 'r') as file:
    reader = csv.reader(file)
    muon_neutrino_mom = []
    for row in reader:
        muon_neutrino_mom.append(float(row[0]))

with open('muon_angle_vals.csv', 'r') as file:
    reader = csv.reader(file)
    muon_angle_wrt_beam_axis = []
    for row in reader:
        muon_angle_wrt_beam_axis.append(float(row[0]))

with open('muon_neutrino_angle_vals.csv', 'r') as file:
    reader = csv.reader(file)
    muon_neutrino_angle_wrt_beam_axis = []
    for row in reader:
        muon_neutrino_angle_wrt_beam_axis.append(float(row[0]))

with open('muon_channel_opening_angle_vals.csv', 'r') as file:
    reader = csv.reader(file)
    muon_channel_opening_angle = []
    for row in reader:
        muon_channel_opening_angle.append(float(row[0]))

with open('muon_channel_avg_opening_angle_vals.csv', 'r') as file:
    reader = csv.reader(file)
    muon_channel_avg_opening_angle = []
    for row in reader:
        muon_channel_avg_opening_angle.append(float(row[0]))
```

```python
#Positron channel data

with open('positron_channel_vals_pion_momentum.csv', 'r') as file:
    reader = csv.reader(file)
    pion_mom_positron_channel_decay = []
    for row in reader:
        pion_mom_positron_channel_decay.append(float(row[0]))

with open('positron_momentum_vals.csv', 'r') as file:
    reader = csv.reader(file)
    positron_mom = []
    for row in reader:
        positron_mom.append(float(row[0]))

with open('electron_neutrino_momentum_vals.csv', 'r') as file:
    reader = csv.reader(file)
    electron_neutrino_mom = []
    for row in reader:
        electron_neutrino_mom.append(float(row[0]))

with open('positron_angle_vals.csv', 'r') as file:
    reader = csv.reader(file)
    positron_angle_wrt_beam_axis = []
    for row in reader:
        positron_angle_wrt_beam_axis.append(float(row[0]))

with open('electron_neutrino_angle_vals.csv', 'r') as file:
    reader = csv.reader(file)
    electron_neutrino_angle_wrt_beam_axis = []
    for row in reader:
        electron_neutrino_angle_wrt_beam_axis.append(float(row[0]))

with open('positron_channel_opening_angle_vals.csv', 'r') as file:
    reader = csv.reader(file)
    positron_channel_opening_angle = []
    for row in reader:
        positron_channel_opening_angle.append(float(row[0]))

with open('positron_channel_avg_opening_angle_vals.csv', 'r') as file:
    reader = csv.reader(file)
    positron_channel_avg_opening_angle = []
    for row in reader:
        positron_channel_avg_opening_angle.append(float(row[0]))
```

```python
plt.hist(pion_mom_muon_channel_decay, bins=300, fill = False, histtype = 'step',density=True)
plt.xlabel('Momentum of pions (MeV)')
plt.ylabel('Probability')
plt.title('Momentum distribution of pions decaying in muon channel')
plt.xlim([0,500])
plt.savefig("Momentum_pions_muon_channel.jpeg")
##files.download("Momentum_pions_muon_channel.jpeg")
```

```python
plt.hist(muon_mom, bins=400, fill = False, histtype = 'step',density=True)
plt.hist(muon_neutrino_mom, bins=100, fill = False, histtype = 'step', edgecolor = 'red',density=True)
plt.xlabel('Momentum (MeV)')
plt.ylabel('Probability')
plt.legend(["Muon","Muon-neutrino"])
plt.title("For Muon Channel")
plt.xlim([0,200])
plt.savefig("Momentum_decayed_products_muon_channel.jpeg")
#files.download("Momentum_decayed_products_muon_channel.jpeg")
```

```python
plt.hist(muon_angle_wrt_beam_axis, bins=100, fill = False, histtype = 'step',density=True)
plt.hist(muon_neutrino_angle_wrt_beam_axis, bins=100, fill = False, histtype = 'step', edgecolor = 'red',density=True)
plt.xlabel('Angle w.r.t. beam axis (in deg)')
plt.ylabel('Probability')
plt.legend(["Muon","Muon-neutrino"])
plt.title("For Muon Channel")
plt.savefig("Decay_Angles.jpeg")
#files.download("Decay_Angles.jpeg")
```

```python
plt.hist(muon_channel_opening_angle, bins=100, fill = False, histtype = 'step',density=True)
plt.xlabel('Opening Angle (deg)')
plt.ylabel('Probability Distribution')
plt.title("For Muon Channel")
plt.savefig("Opening_Angle.jpeg")
#files.download("Opening_Angle.jpeg")
```

```python
A  = range(0,10*len(muon_channel_avg_opening_angle),10)
plt.plot(A, muon_channel_avg_opening_angle,".-")
plt.xlabel("Pion momentum (MeV)")
plt.ylabel("Average Opening Angle (Degrees)")
plt.title("For Muon Channel")
plt.savefig("Avg_Opening_Angle_Muon_Channel.jpeg")
#files.download("Avg_Opening_Angle_Muon_Channel.jpeg")
```

```python
plt.hist(pion_mom_positron_channel_decay, bins=100, fill = False, histtype = 'step',density=True)
plt.xlabel('Momentum of pions (MeV)')
plt.ylabel('Probability')
plt.title('Momentum distribution of pions decaying in positron channel')
plt.xlim([0,500])
plt.savefig("Pion_mom_positron_channel.jpeg")
#files.download("Pion_mom_positron_channel.jpeg")
```

```python
plt.hist(positron_mom, bins=300, fill = False, histtype = 'step',density=True)
plt.hist(electron_neutrino_mom, bins=100, fill = False, histtype = 'step', edgecolor = 'red',density=True)
plt.xlabel('Momentum (MeV)')
plt.ylabel('Probability')
plt.xlim([0,200])
plt.legend(["positron","electron-neutrino"])
plt.title("For Positron Channel")
plt.savefig("Decay_mom_positron_channel.jpeg")
#files.download("Decay_mom_positron_channel.jpeg")
```

```python
plt.hist(positron_angle_wrt_beam_axis, bins=100, fill = False, histtype = 'step',density=True)
plt.hist(electron_neutrino_angle_wrt_beam_axis, bins=100, fill = False, histtype = 'step', edgecolor = 'red',density=True)
plt.xlabel('Angle w.r.t. beam axis (in deg)')
plt.ylabel('Probability')
plt.legend(["positron","electron-neutrino"])
plt.title("For Positron Channel")
plt.savefig("Angles_positron_channel.jpeg")
#files.download("Angles_positron_channel.jpeg")
```

```python
plt.hist(positron_channel_opening_angle, bins=100, fill = False, histtype = 'step',density=True)
plt.xlabel('Opening Angle (deg)')
plt.ylabel('Probability')
plt.title("For Positron Channel")
plt.savefig("Opening_Angles_positron_channel.jpeg")
#files.download("Opening_Angles_positron_channel.jpeg")
```
